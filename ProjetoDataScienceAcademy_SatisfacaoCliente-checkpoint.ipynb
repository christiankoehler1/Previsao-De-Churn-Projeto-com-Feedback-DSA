```python
# Separando os dados das duas classes para posteriormente balancear os dados - os dados são balanceados para evitar que o 
# algoritmo aprenda mais com 1 resultado do que com o outro.

X = clientes_df.loc[:,'var3':'var38']
y = clientes_df.loc[:,'TARGET']
```

```python
# Verificando o tipo de dado
type(y)
```

```python
# Transformando TARGET (y) em um Data Frame
y = pd.DataFrame(y)
type(y)
```

```python
# Método escolhido foi o Undersampling NearMiss
# Carregando o pacote
from imblearn.under_sampling import NearMiss
```

**NearMiss – É um algoritmo de undersampling que consiste em reduzir de forma aleatória os exemplos da classe majoritária, 
porém ele seleciona os exemplos com base na distância.**

```python
# Balanceando os dados
nr = NearMiss()
X,y = nr.fit_resample(X,y)
```

```python
# Verificando o tamanho de X (Atributos)
X
```

```python
# Verificando o tamanho de y (Target)
y
```

```python
# Separando os dados em treino e teste

# Importando os módulos 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Definindo o tamanho da amostra - divisão do dataset em 70/30 (treino/teste)
test_size = 0.30

# Garantindo que os resultados possam ser reproduzidos
# É importante poiss assim é possível comparar a acurácia com outros algoritmos de Machine Learning.
seed = 7

# Criando os conjuntos de treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X,y,test_size = test_size, random_state = seed)

# Criando o modelo
modelo = LogisticRegression()

# Treinando o modelo
modelo.fit(X_treino, y_treino)

# Predição do modelo
y_pred = modelo.predict(X_teste)

# Score do modelo nos dados de teste
resultado = modelo.score(X_teste,y_teste)
print('Acurácia nos dados de testes: %.3f%%' %(resultado*100))

# Avaliação do resultado pela Matriz de Confusão
print(confusion_matrix(y_teste, y_pred))
```

## Melhorando a Acurácia do modelo

```python
# Normalizando os dados - deixando as variáveis na mesma escala (entre 0 e 1)

# Importar o módulo
from sklearn.preprocessing import MinMaxScaler

# Gerando a nova escala
escala = MinMaxScaler(feature_range = (0,1))
redimensionaX = escala.fit_transform(X)

# Verificando os dados após normalização
print('Dados Originais: n/n/', X.head(5))
print('\n Dados Normalizados: n/n/', redimensionaX[0:5,:])
```

```python
# Padronizando os dados - para transformar os atributos numa distribuição normal (média 0 e desvio padrão 1)

# Importar o módulo
from sklearn.preprocessing import StandardScaler

# Gerando a nova distribuição
escala = StandardScaler().fit(X)
padraoX = escala.transform(X)

# Verificando os dados após normalização
print('Dados Originais: n/n/', X.head(5))
print('\n Dados Normalizados: n/n/', padraoX[0:5,:])
```

```python
# Separando os dados em treino e teste

# Importando os módulos 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Definindo o tamanho da amostra - divisão do dataset em 70/30 (treino/teste)
test_size = 0.30

# Garantindo que os resultados possam ser reproduzidos
# É importante poiss assim é possível comparar a acurácia com outros algoritmos de Machine Learning.
seed = 8

# Criando os conjuntos de treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X,y,test_size = test_size, random_state = seed)

# Criando o modelo
modelo = LogisticRegression()

# Treinando o modelo
modelo.fit(X_treino, y_treino)

# Predição do modelo
y_pred = modelo.predict(X_teste)

# Score do modelo nos dados de teste
resultado = modelo.score(X_teste,y_teste)
print('Acurácia nos dados de testes: %.3f%%' %(resultado*100))

# Avaliação do resultado pela Matriz de Confusão
print(confusion_matrix(y_teste, y_pred))
```
